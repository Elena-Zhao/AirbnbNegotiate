# Calendar related attributes

```python
def parse_calendar(list_calendar):
    date = []
    price_USD = []
    availability = []
    for month in list_calendar['calendar_months']:
        for day in month['days']:
            date.append(day['date'])
            price_USD.append(day['price']['native_price'])
            availability.append(day['available'])
    return pd.DataFrame({'date':date, 'price_USD': price_USD, 'availability':availability})

def find_orphans(cal, length):
    orphan_start = []
    orphan_end = []
    orphan_price = []

    index = 0
    while index < len(cal['availability']) - length:
        count = 0
        start = index
        while cal['availability'][index] == True and count < length:
            count += 1
            if count >= length and cal['availability'][index + 1] == False:
                orphan_start.append(cal['date'][start])
                orphan_end.append(cal['date'][index])
                if length == 1:
                    orphan_price.append(cal['price_USD'][start])
                else:
                    orphan_price.append(np.mean(cal['price_USD'][start: index]))
                break
            index += 1
        index += 1
    if len(orphan_start) > 0:
        return pd.DataFrame({'Start': orphan_start, 'End': orphan_end, 'Price': orphan_price})
    else:
        return []

def listings_price_var(calendars):
    listings_var = {}
    for i in calendars.keys():
        listings_var[i] = np.var(parse_calendar(calendars[i])['price_USD'])
    return listings_var
```

```python
import datetime
def days_in_advance(date):
    date = datetime.datetime.strptime(date, '%Y-%m-%d')
    days_in_advance = (date - datetime.datetime.today()).days
    return days_in_advance
```

```python
'''
differnce between this and find_orphans(cal, length):
Incooperate all necessary infomation for assiging to different buckets

'''

def find_orphans_2(length, listing_id, parsed_calendars = parsed_calendars, listings = listings):
    cal = parsed_calendars[listing_id]
    orphan_start = []
    orphan_end = []
    orphan_price = []

    index = 0
    while index < len(cal['availability']) - length:
        count = 0
        start = index
        while cal['availability'][index] == True and count < length:
            count += 1
            if count >= length and cal['availability'][index + 1] == False:
                orphan_start.append(cal['date'][start])
                orphan_end.append(cal['date'][index])
                if length == 1:
                    orphan_price.append(cal['price_USD'][start])
                else:
                    orphan_price.append(np.mean(cal['price_USD'][start: index]))
                break
            index += 1
        index += 1
    if len(orphan_start) > 0:
        host_id = listings['host_id'][listings['id'] == listing_id].tolist() * len(orphan_start)
        days_advance = [days_in_advance(date) for date in orphan_start]
        return pd.DataFrame({'host_id' : host_id,'listing_id': listing_id, 'Start': orphan_start, 
                             'End': orphan_end, 'Price': orphan_price, 'days_advance': days_advance})
    else:
        return []
```

```python
def orphan_bucket(orphan_bucket_num, listings = listings):
    orphan_bucket = []
    for i in range(1, orphan_bucket_num + 1):
        orphan_list = []
        for id in parsed_calendars.keys():
            # Whether the length satisfys the listing's minimun nights
            if i >= listings[listings['id'] == id]['minimum_nights'].to_dict().values()[0]:
                orphan = find_orphans_2(i, id)
                if len(orphan) > 0:
                    orphan_list.append(orphan)

        orphan_df = pd.concat(orphan_list)
        orphan_bucket.append(orphan_df)
    return orphan_bucket
```

```python
'''
Hi, my name is Michele  and I am interested in staying at your place for 1 day from 3/20/16 to 3/21/16. 
The calendar is open on these days and I am wondering if it is still available? 

A little bit more about myself: I am a graduate student from Milan, Italy and study Communication Design.
I will visit Boston a few times this year to work on a project with a local school. 
I am fond of exploring cultures and visual arts around the world.

I am wondering if you would be willing to lower the rate to $85/night? The trip is really soon and my schedule matches perfectly with the opening in your calendar. So I would be willing to make the booking immediately if the price is ok with you.

'''

import datetime
def calculate_duration(start, end):
    start = datetime.datetime.strptime(start, '%Y-%m-%d')
    end = datetime.datetime.strptime(end, '%Y-%m-%d')

    days_in_advance = (end - start).days
    return days_in_advance

def generateMessage(host_name, start, end, price, num_guests):
    #this is a list storing all reasons for each 
    days_advance = (datetime.datetime.strptime(start, '%Y-%m-%d') - datetime.datetime.today()).days
    if days_advance < 7:
        reason = " Since the trip is coming up very soon, and I found my schedule matches perfectly with the opening in your calendar,"
    else: 
        reason = " Because my schedule match perfectly with the opening in your calendar,"

    duration = calculate_duration(start, end)
    if duration == 1:
        duration = str(duration) + ' day'
    else:
        duration = str(duration) + ' days'
        
    message_1 = 'Hi {}, my name is Michele and I am interested in staying at your place for {}, from {} to {}. The calendar is open on these days and I am wondering if it is still available? '.format(host_name ,duration, start , end)
    
    message_2 = 'A little bit more about myself: I am a graduate student from Milan, Italy and study Communication Design. I am fond of exploring cultures and visual arts around the world.'
    
    message_3 = '{} I am wondering if you would be willing to lower the rate to ${}/night? I would be willing to make the booking immediately if the price is ok with you.'.format(reason,price)
        
    if(num_guests==1 or num_guests==0 or np.isnan(num_guests)):
        return message_1 + message_2 + message_3
    elif (num_guests ==2 ):
        addition_two_persons = 'I will visit Boston a few times this year to work on a project with a local school. I will be there with a collegue of mine. '
        return message_1 + message_2 + addition_two_persons + message_3
    else: 
        addition_n_persons = 'I will visit Boston a few times this year to work on a project with a local school. I will be there with {} collegues of mine. '.format(int(num_guests - 1))
        return message_1 + message_2 + addition_n_persons + message_3
```
